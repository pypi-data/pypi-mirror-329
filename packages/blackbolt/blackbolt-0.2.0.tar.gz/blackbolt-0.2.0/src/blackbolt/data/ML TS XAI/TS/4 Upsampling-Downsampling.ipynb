```python
import pandas as pd

series1 = pd.read_csv(url, header=0, index_col=0)
series1.index = pd.to_datetime('190' + series1.index, format='%Y-%m')
print(series1)
```

```python
upsampled = series1.resample('D')
interpolated = upsampled.interpolate(method='linear')
print(interpolated.head(32))
```

```python
interpolated = upsampled.interpolate(method='spline',order=2)
print(interpolated.head(32))
```

```python
interpolated.plot()
plt.show()
```

```python
resample = series1.resample('Q')
quarterly_mean_sales = resample.mean()
print(quarterly_mean_sales.head())
```

```python
quarterly_mean_sales.plot()
plt.show()
```

```python
resample = series1.resample('A')
quarterly_mean_sales = resample.sum()
print(quarterly_mean_sales.head())
```

```python
quarterly_mean_sales.plot()
plt.show()
```
