```python
import numpy as np
import pandas as pd
import pytae as pt
import seaborn as sns

penguins = sns.load_dataset('penguins')
```

```python
(penguins
 .select(['body_mass_g','bill_length_mm','species','island','sex'])
 .qry({'species':'Adelie'}) #can use pandas .query but this is more flexible for all 'and' queries!
 .agg_df()
)
#for all practical purpose majority of query is 'and' by nature.
```

```python
(penguins
  .agg_df()
  .qry({'species':'Adelie','body_mass_g':74125}) # == is by default
)
```

```python
(penguins
  .agg_df()
  .qry({'body_mass_g':[74125,75000,76000,90300]}) # also accepts list
)
```

```python
(penguins
  .agg_df()
  .qry({'species':['Adelie','Gentoo'],'body_mass_g':"> 81500"}) 
 
 #conditions can be passed under quotation; 
 #for numeric whitespaces will be taken care of 
)
```

```python
(penguins
  .agg_df()
  .qry({'species':">=Adelie",'flipper_length_mm':"==4118"}) 
)
```

```python
(penguins
  .agg_df()
  .qry({'species':">=Adelie",'flipper_length_mm':"<12000",'flipper_length_mm':">5000"}) 
) 

#since input is dict it will only recognize the last key if keys are duplicated;
#to overcome this limitation either do .qry again or also use 
# ( >; [ >=; ) <; ]<= the standard mathemetical notation for between x and y kind of queries which needs 
# to be performed on same col
```

```python
(penguins
  .agg_df()
  .qry({'body_mass_g': '[89100,120000)'}) # >= 89100 and <120000
)
```

```python
(penguins
  .agg_df()
  .qry({'island': '(Biscoe,Dream]'}) #all island greater than Biscoe and upto Dream
)
```

```python
(penguins
  .agg_df()
  .qry({'body_mass_g': [89100,120000]}) 
 )
 # note without quotation it is correctly treated as list and not closed ended brackets!
```
